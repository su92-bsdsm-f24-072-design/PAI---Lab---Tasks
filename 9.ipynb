```python
import cv2
import matplotlib.pyplot as plt

# Read image (BGR format)
img = cv2.imread('g4g.png')

if img is None:
    print("Image not loaded")
else:
    # Convert BGR to YCrCb
    ycrcb = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)

    # Convert YCrCb back to RGB for correct display
    rgb_image = cv2.cvtColor(ycrcb, cv2.COLOR_YCrCb2RGB)

    plt.imshow(rgb_image)
    plt.axis('off')
    plt.title("YCrCb Image (Displayed Correctly)")
    plt.show()
```

```python
# Python program to read image as HSV and display in Jupyter Notebook

import cv2
import matplotlib.pyplot as plt

# Read the image
img = cv2.imread('g4g.png')

# Check if image loaded correctly
if img is None:
    print("Image not loaded. Check filename or path!")
else:
    # Convert to HSV
    hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    # Convert HSV to RGB for proper matplotlib display
    hsv_rgb = cv2.cvtColor(hsv_img, cv2.COLOR_HSV2RGB)

    # Display the image inline
    plt.figure(figsize=(6,6))
    plt.imshow(hsv_rgb)
    plt.title("HSV Image")
    plt.axis('off')
    plt.show()
```

```python
# Python program to read image as LAB color space and display in Jupyter Notebook

import cv2
import matplotlib.pyplot as plt

# Read the image (BGR format)
img = cv2.imread('g4g.png')

# Check if image loaded correctly
if img is None:
    print("Image not loaded. Check the filename or path!")
else:
    # Convert to LAB color space
    lab_img = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)

    # Convert LAB to RGB for correct display in matplotlib
    lab_rgb = cv2.cvtColor(lab_img, cv2.COLOR_LAB2RGB)

    # Display the image inline
    plt.figure(figsize=(6,6))
    plt.imshow(lab_rgb)
    plt.title("LAB Image")
    plt.axis('off')
    plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Read the image
img = cv2.imread('g4g.png')

# Check if image loaded correctly
if img is None:
    print("Image not loaded. Check filename or path!")
else:
    # Convert to grayscale first (recommended for edge detection)
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Apply Laplacian edge detection
    laplacian = cv2.Laplacian(gray_img, cv2.CV_64F)

    # Convert float64 to uint8 for display
    laplacian_display = cv2.convertScaleAbs(laplacian)

    # Display the edge map inline
    plt.figure(figsize=(6,6))
    plt.imshow(laplacian_display, cmap='gray')
    plt.title("Edge Map (Laplacian)")
    plt.axis('off')
    plt.show()
```
