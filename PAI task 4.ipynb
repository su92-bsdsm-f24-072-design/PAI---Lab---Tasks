```python
def print_board(board, n):
    """
    This function prints the chessboard in a readable format.
    Q represents a Queen and. represents an empty space.
    """
    print("\n Solution Board:\n")
    for row in board:
        print(" ".join(row))
    print("\n")

def is_safe(board, row, col, n):
    """
    This function checks whether placing a queen at (row,col)
    is safe or not.
    """
    for i in range(row):
        if board[i][col] == "Q":
            return False
    i, j = row, col
    while i >= 0 and j >= 0:
        if board[i][j] == "Q":
            return False
        i -= 1
        j -= 1
    i, j = row, col
    while i >= 0 and j < n:
        if board[i][j] == "Q":
            return False
        i -= 1
        j += 1
    return True

def solve_nqueens(board, row, n):
    """
    Backtracking function to solve N-Queens problem.
    Places queens row by row.
    """
    if row == n:
        return True
    for col in range(n):

        if is_safe(board, row, col, n):

            board[row][col] = "Q"

            if solve_nqueens(board, row + 1, n):
                return True
            board[row][col] = "."
    return False
    
def main():
    print(" N-Queens Problem Solver (Dynamic Input)\n")
    n = int(input("Enter the value of N (e.g., 4, 8): "))
    board = [["." for _ in range(n)] for _ in range(n)]
    if solve_nqueens(board, 0, n):
        print_board(board, n)
    else:
        print("No solution exists for this value of N.")
main()
```
