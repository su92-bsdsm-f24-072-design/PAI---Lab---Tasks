```python
import cv2
import matplotlib.pyplot as plt

# Load image
img = cv2.imread(r"C:\Users\Fakhar Trader\Desktop\PAI(lab)\task 5\flower.jpg")

# If OpenCV fails, try matplotlib
if img is None:
    import matplotlib.image as mpimg
    img = mpimg.imread(r"C:\Users\Fakhar Trader\Desktop\PAI(lab)\task 5\flower.jpg")
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

# Grayscale + Histogram
g = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.subplot(121)
plt.imshow(g, cmap='gray')
plt.axis("off")
plt.title("Grayscale")

plt.subplot(122)
plt.hist(g.ravel(),256,[0,256],color='k')
plt.title("Gray Histogram")
plt.show()

# RGB Histograms
for i,c in enumerate(('r','g','b')):
    plt.plot(cv2.calcHist([img],[i],None,[256],[0,256]), color=c)

plt.title("RGB Histograms")
plt.xlabel("Intensity")
plt.ylabel("Frequency")
plt.show()
```

```python
img = cv2.imread('flower.jpg', 0)
histg = cv2.calcHist([img], [0], None, [256], [0, 256])

# Plot histogram
plt.plot(histg)
plt.title("Histogram using OpenCV calcHist()")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('ex.jpg')

# check image
if img is None:
    raise FileNotFoundError("Image not found. Check the file path!")

# colors for channels
colors = ('b', 'g', 'r')

for i, col in enumerate(colors):
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist, color=col)
    plt.xlim([0, 256])

plt.title("RGB Color Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread(".png")  # use the correct filename

# Safety check
if img is None:
    raise FileNotFoundError("Image not found. Check filename/path!")

# Colors for RGB channels
colors = ('b','g','r')

# Plot histograms
for i, col in enumerate(colors):
    hist = cv2.calcHist([img], [i], None, [256], [0,256])
    plt.plot(hist, color=col)
    plt.xlim([0,256])

plt.title("RGB Color Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()
```
