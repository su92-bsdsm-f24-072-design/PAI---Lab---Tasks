# Water Jug Problem:

```python
def water_jug_dfs(capacity1, capacity2, target):
    stack = [((0, 0), [])]
    visited = set()

    while stack:
        (x, y), path = stack.pop()
        if (x, y) in visited:
            continue

        visited.add((x, y))
        path = path + [((x, y), "Current State")]

        if x == target or y == target:
            print("\nGoal Reached Successfully!\n")
            for step in path:
                print(step)
            return

        next_states = []
        next_states.append(((capacity1, y), "Fill Jug 1"))
        next_states.append(((x, capacity2), "Fill Jug 2"))
        next_states.append(((0, y),"Empty Jug 1"))
        next_states.append(((x, 0),"Empty Jug 2"))

        transfer = min(x, capacity2 - y)
        next_states.append(((x - transfer,y + transfer),
                            "Pour Jug 1 -> Jug 2"))
        transfer = min(y, capacity1 - x)
        next_states.append(((x + transfer,y - transfer),
                            "Pour Jug 2 -> Jug 1"))

        for state, rule in next_states:
            if state not in visited:
                stack.append((state, path + [(state, rule)]))
    print("No solution found")

water_jug_dfs(4,3,2)
```
