```python
import cv2
import matplotlib.pyplot as plt

src = cv2.imread("logo.png")

if src is None:
    print("Image not loaded")
else:
    gray_image = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)

    plt.imshow(gray_image, cmap="gray")
    plt.axis("off")
    plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

src = cv2.imread("logo.png")

if src is None:
    print("Image not loaded")
else:
    hsv_image = cv2.cvtColor(src, cv2.COLOR_BGR2HSV)
    hsv_to_rgb = cv2.cvtColor(hsv_image, cv2.COLOR_HSV2RGB)

    plt.imshow(hsv_to_rgb)
    plt.axis("off")
    plt.title("HSV Image (Displayed Correctly)")
    plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
src = cv2.imread(r'logo.png')

# Convert from BGR to RGB
rgb_image = cv2.cvtColor(src, cv2.COLOR_BGR2RGB)

# Display with Matplotlib
plt.imshow(rgb_image)
plt.title("RGB Image for Matplotlib")
plt.axis('off')
plt.show()
```
