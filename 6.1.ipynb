# Co-ordinates of Contours

```python
import numpy as np
import cv2

# Font for text
font = cv2.FONT_HERSHEY_COMPLEX

# Load image (check if file exists and path is correct)
img_color = cv2.imread('test.jpg', cv2.IMREAD_COLOR)
img_gray = cv2.imread('test.jpg', cv2.IMREAD_GRAYSCALE)

if img_color is None or img_gray is None:
    print("Error: Image not loaded. Check the path and filename!")
    exit()

# Binarize image
_, threshold = cv2.threshold(img_gray, 110, 255, cv2.THRESH_BINARY)

# Find contours
contours, _ = cv2.findContours(threshold, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Create a copy to draw contours
img_drawn = img_color.copy()

for cnt in contours:
    # Approximate contour to polygon
    approx = cv2.approxPolyDP(cnt, 0.009 * cv2.arcLength(cnt, True), True)
    
    # Draw contour
    cv2.drawContours(img_drawn, [approx], 0, (0, 0, 255), 2)
    
    # Flatten points and label
    n = approx.ravel()
    for i in range(0, len(n), 2):
        x, y = n[i], n[i+1]
        coord = f"{x} {y}"
        if i == 0:
            cv2.putText(img_drawn, "Arrow tip", (x, y-5), font, 0.5, (255, 0, 0), 1)
        else:
            cv2.putText(img_drawn, coord, (x, y-5), font, 0.5, (0, 255, 0), 1)

# Show result
cv2.imshow('Contours with Coordinates', img_drawn)

# Wait until a key is pressed, then close
cv2.waitKey(0)
cv2.destroyAllWindows()
```
