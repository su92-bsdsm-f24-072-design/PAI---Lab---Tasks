```python
import cv2
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('taj.jpg')

# Apply bilateral filter
bilateral = cv2.bilateralFilter(img, 15, 75, 75)

# Save the output
cv2.imwrite('taj_bilateral.jpg', bilateral)

# Convert BGR to RGB for matplotlib
bilateral_rgb = cv2.cvtColor(bilateral, cv2.COLOR_BGR2RGB)

# Show the image inline
plt.imshow(bilateral_rgb)
plt.axis('off')
plt.title("Bilateral Filtered Image")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Read the image
img = cv2.imread('taj.jpg')

if img is None:
    print("Image not loaded! Check the path/filename")
else:
    # --- Apply filters ---
    # 1. Bilateral Filter
    bilateral = cv2.bilateralFilter(img, d=15, sigmaColor=75, sigmaSpace=75)

    # 2. Average Filter (Blur)
    average = cv2.blur(img, ksize=(15, 15))

    # 3. Median Filter
    median = cv2.medianBlur(img, ksize=15)

    # Convert BGR to RGB for matplotlib
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    bilateral_rgb = cv2.cvtColor(bilateral, cv2.COLOR_BGR2RGB)
    average_rgb = cv2.cvtColor(average, cv2.COLOR_BGR2RGB)
    median_rgb = cv2.cvtColor(median, cv2.COLOR_BGR2RGB)

    # --- Display images side by side ---
    titles = ['Original', 'Bilateral', 'Average', 'Median']
    images = [img_rgb, bilateral_rgb, average_rgb, median_rgb]

    plt.figure(figsize=(16, 8))
    for i in range(4):
        plt.subplot(1, 4, i+1)
        plt.imshow(images[i])
        plt.title(titles[i])
        plt.axis('off')
    plt.show()
```
