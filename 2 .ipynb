# Method 2: Using the cv2.imread() function with flag=zero

```python
import cv2
from matplotlib import pyplot as plt

image_path = 'cats_dogs.jpg'
image = cv2.imread(image_path)

if image is None:
    print("Error: Image not loaded. Check file path and name!")
else:
    print("Image loaded successfully!")
    resized_image = cv2.resize(image, (1900, 800))
    resized_image_rgb = cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)
    plt.imshow(resized_image_rgb)
    plt.title("Resized Image")
    plt.axis('off')
    plt.show()
```

# Method 3.2: Using the pixel manipulation (Average method)

```python
import cv2
import os

image_path = r"C:\Users\Fakhar Trader\Desktop\PAI(lab)\task 5\tomatoes.webp"
if not os.path.exists(image_path):
    print(" File does not exist. Check path.")
else:
    img = cv2.imread(image_path)
    if img is None:
        print("Image not loaded. Convert to JPG and try again.")
    else:
        print("Image loaded successfully")
        rows, cols = img.shape[:2]
        for i in range(rows):
            for j in range(cols):
                gray = int((int(img[i, j, 0]) +
                            int(img[i, j, 1]) +
                            int(img[i, j, 2])) / 3)
                img[i, j] = [gray, gray, gray]

        cv2.imshow('Grayscale Image (Average)', img)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
```
