```python
# folder_structure.py
import os

def print_tree(startpath, prefix=""):
    """Print folder structure recursively with indentation"""
    for index, item in enumerate(sorted(os.listdir(startpath))):
        path = os.path.join(startpath, item)
        is_last = index == len(os.listdir(startpath)) - 1
        connector = "└── " if is_last else "├── "
        print(prefix + connector + item)
        if os.path.isdir(path):
            extension = "    " if is_last else "│   "
            print_tree(path, prefix + extension)

if __name__ == "__main__":
    # Set your project folder here
    project_folder = r"C:\Users\Fakhar Trader\Desktop\AnimalHerdDetection"
    print(project_folder)
    print_tree(project_folder)
```
