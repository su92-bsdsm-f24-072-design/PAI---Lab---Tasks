```python
import cv2
import matplotlib.pyplot as plt

# Load image
image = cv2.imread("logo.png")

# Check if image loaded
if image is None:
    print("Image not loaded! Check filename/path")
else:
    # Create different borders
    border_reflect = cv2.copyMakeBorder(image, 50, 50, 50, 50, cv2.BORDER_REFLECT)
    border_reflect_101 = cv2.copyMakeBorder(image, 50, 50, 50, 50, cv2.BORDER_REFLECT_101)
    border_replicate = cv2.copyMakeBorder(image, 50, 50, 50, 50, cv2.BORDER_REPLICATE)

    # Convert BGR → RGB for matplotlib
    border_reflect_rgb = cv2.cvtColor(border_reflect, cv2.COLOR_BGR2RGB)
    border_reflect_101_rgb = cv2.cvtColor(border_reflect_101, cv2.COLOR_BGR2RGB)
    border_replicate_rgb = cv2.cvtColor(border_replicate, cv2.COLOR_BGR2RGB)

    # Display images inline
    plt.figure(figsize=(15,5))

    plt.subplot(1,3,1)
    plt.imshow(border_reflect_rgb)
    plt.title("Border Reflect")
    plt.axis('off')

    plt.subplot(1,3,2)
    plt.imshow(border_reflect_101_rgb)
    plt.title("Border Reflect 101")
    plt.axis('off')

    plt.subplot(1,3,3)
    plt.imshow(border_replicate_rgb)
    plt.title("Border Replicate")
    plt.axis('off')

    plt.show()
```
